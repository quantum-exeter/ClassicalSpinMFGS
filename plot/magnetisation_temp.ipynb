## Classical Magnetisation vs. Temperature Curves ##

```python
import numpy as np
import numpy.random as rn
import matplotlib
import matplotlib.pyplot as plt
import proplot as pplt
import pandas as pd
```

```python
data = np.array(pd.read_csv(r'C://Users//crh222//Data//Classical//test.csv'))
```

```python
fig, ax = pplt.subplots(figsize=(6.5, 4.5))  #size in inches

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

ax.set_xlabel("$2 k_B T/\hbar \omega_L$", fontsize=12)
ax.set_ylabel("$S_z$", fontsize=12)

ax.set_xlim(0.01, 1000)

##### Data ######
ax.semilogx(data[:,0], data[:,1], 'k-')

##### Plot ######
plt.grid(which='both', lw='0.2')
```
